```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data/2021-2022 NBA Player Stats.csv", sep=";")
df
```

```python
df.columns
```

Rk : Rank
Player : Player's name
Pos : Position
Age : Player's age
Tm : Team
G : Games played
GS : Games started
MP : Minutes played per game
FG : Field goals per game
FGA : Field goal attempts per game
FG% : Fiel goal percentage
3P : 3-point field goals per game
3PA : 3-point field goal attempts per game
3P% : 3-point field goal percentage
2P : 2-point field goals per game
2PA : 2-point field goal attempts per game
2P% : 2-point field goal percentage
eFG% : Effective field goal percentage
FT : Free throws per game
FTA : Free throw attempts per game
FT% : Free throw percentage
ORB : Offensive rebounds per game
DRB : Defensive rebounds per game
TRB : Total rebounds per game
AST : Assists per game
STL : Steals per game
BLK : Blocks per game
TOV : Turnovers per game
PF : Personal fouls per game
PTS : Points per game

```python
df = df.drop(columns=['Rk'])
df
```

```python
df.groupby("Tm").mean()
```

```python
df.groupby("Pos").mean()
```

```python
df["Player"].count()
```

```python
Efficiency = df.PTS + df.TRB + df.AST + df.STL + df.BLK - (df.FGA - df.FG) - (df.FTA - df.FT) - df.TOV
df.insert(29, "EFF", Efficiency)
df.head(10)
```

```python
players_C = df[df.Pos == "C"]
players_PG = df[df.Pos == "PG"]
players_SG = df[df.Pos == "SG"]
players_PF = df[df.Pos == "PF"]
players_SF = df[df.Pos == "SF"]

players_PG.head(10)
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
sns.scatterplot('Age','PTS',hue='Pos',data=df)
plt.xlabel('Age')
plt.ylabel('Points per Game')
plt.title('Age vs Points per Game')
plt.text(df.Age[df.Player=="LeBron James"]+0.3,df.PTS[df.Player=="LeBron James"]+0.3,"Lebron James", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(df.Age[df.Player=="DeMar DeRozan"]+0.3,df.PTS[df.Player=="DeMar DeRozan"]+0.3,"DeMar DeRozan", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(df.Age[df.Player=="Kevin Durant"]+0.3,df.PTS[df.Player=="Kevin Durant"]+0.3,"Kevin Durant", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(df.Age[df.Player=="Stephen Curry"]+0.3,df.PTS[df.Player=="Stephen Curry"]+0.3,"Stephen Curry", color='black',bbox=dict(facecolor='yellow',alpha=0.5))

plt.legend(title='Position');
```

```python
Cs = pd.DataFrame(players_C.Player[(players_C.Age<=23)&(players_C.PTS>=15)&(players_C.EFF>=12)]).rename(columns={"Player": "Centers"}).reset_index(drop=True)
PGs = pd.DataFrame(players_PG.Player[(players_PG.Age<=23)&(players_PG.PTS>=15)&(players_PG.EFF>=12)]).rename(columns={"Player": "Point Guards"}).reset_index(drop=True)
SGs = pd.DataFrame(players_SG.Player[(players_SG.Age<=23)&(players_SG.PTS>=15)&(players_SG.EFF>=12)]).rename(columns={"Player": "Shooting Guards"}).reset_index(drop=True)
PFs = pd.DataFrame(players_PF.Player[(players_PF.Age<=23)&(players_PF.PTS>=15)&(players_PF.EFF>=12)]).rename(columns={"Player": "Power Forwards"}).reset_index(drop=True)
SFs = pd.DataFrame(players_SF.Player[(players_SF.Age<=23)&(players_SF.PTS>=15)&(players_SF.EFF>=12)]).rename(columns={"Player": "Small Forwards"}).reset_index(drop=True)

future = pd.concat([Cs, PGs, SGs, PFs, SFs], axis=1, join="outer")
future
```
