```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df = pd.read_csv("data/2021-2022 NBA Player Stats.csv", sep=";")
df
```

```python
df.columns
```

Pos : Position
Age : Player's age
Tm : Team
G : Games played
GS : Games started

FG : Field goals per game
FGA : Field goal attempts per game
FG% : Fiel goal percentage
3P : 3-point field goals per game
3PA : 3-point field goal attempts per game
3P% : 3-point field goal percentage
2P : 2-point field goals per game

```python
df = df.drop(columns=['Rk'])
df
```

```python
df.groupby("Tm").mean().round(2)
```

```python
df.groupby("Pos").mean().round(2)
```

```python
df["Player"].count()
```

```python
Efficiency = df.PTS + df.TRB + df.AST + df.STL + df.BLK - (df.FGA - df.FG) - (df.FTA - df.FT) - df.TOV
df.insert(29, "EFF", Efficiency)
df.head(10)
```

```python
players_C = df[df.Pos == "C"]
players_PG = df[df.Pos == "PG"]
players_SG = df[df.Pos == "SG"]
players_PF = df[df.Pos == "PF"]
players_SF = df[df.Pos == "SF"]

players_PG.head(10)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True);
```
